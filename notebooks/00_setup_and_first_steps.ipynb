## Learning goals

```python
import os
import sys
from pathlib import Path

# Keep library caches inside the project so the notebook needs no administrator access.
project_root = Path.cwd().parent if Path.cwd().name == "notebooks" else Path.cwd()
matplotlib_cache = project_root / ".cache" / "matplotlib"
matplotlib_cache.mkdir(parents=True, exist_ok=True)
os.environ.setdefault("MPLCONFIGDIR", str(matplotlib_cache))

import matplotlib
import numpy as np
import pandas as pd
import seaborn as sns
import sklearn

print("Python executable:", sys.executable)
print("Python version:", sys.version.split()[0])
print("pandas:", pd.__version__)
print("NumPy:", np.__version__)
print("scikit-learn:", sklearn.__version__)
print("Matplotlib:", matplotlib.__version__)
print("Seaborn:", sns.__version__)
```

## Why check `sys.executable`?

A computer can contain several Python installations. `sys.executable` tells us exactly which interpreter is executing this notebook. Its path should contain this project's `.venv`.

```python
# The notebook is inside notebooks/, so its parent directory is the project root.
raw_data_dir = project_root / "data" / "raw"

print("Project root:", project_root.resolve())
print("Raw data folder:", raw_data_dir.resolve())
print("CSV files currently present:", [path.name for path in raw_data_dir.glob("*.csv")])
```

## The first five dataset checks

Once a CSV is placed in `data/raw/`, these checks answer:

1. What are the rows and columns?
2. What does a sample look like?
3. Which data types did pandas infer?
4. Where are values missing?
5. Are complete rows duplicated?

```python
csv_files = list(raw_data_dir.glob("*.csv"))

if not csv_files:
    print("No CSV found yet. Setup is working; the next lesson will download the official crop data.")
else:
    data = pd.read_csv(csv_files[0])
    display(data.head())
    print("Shape (rows, columns):", data.shape)
    display(data.dtypes.rename("dtype"))
    display(data.isna().sum().sort_values(ascending=False).rename("missing_values"))
    print("Duplicate rows:", data.duplicated().sum())
```

## Interview note

Do not begin model training immediately after downloading a CSV. First verify provenance, units, schema, missing values, duplicates, target meaning, and possible data leakage.
